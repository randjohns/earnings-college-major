# College Job Outcomes Analysis

College major type and gender are two frequently discussed and debated factors that influence future job success as measured by salary, number of jobs, and FT/PT status.

The present analysis is based on a dataset of job outcomes published by the American Community Survey company. The data includes information on students who graduated from college between 2010 and 2012. FiveThirtyEight cleaned the dataset and released it on their Github repository.

The dataset includes the following information formatted into columns:

- Rank - Rank by median earnings (the dataset is ordered by this column).
- Major_code - Major code.
- Major - Major description.
- Major_category - Category of major.
- Total - Total number of people with major.
- Sample_size - Sample size (unweighted) of full-time.
- Men - Male graduates.
- Women - Female graduates.
- ShareWomen - Women as share of total.
- Employed - Number employed.
- Median - Median salary of full-time, year-round workers.
- Low_wage_jobs - Number in low-wage service jobs.
- Full_time - Number employed 35 hours or more.
- Part_time - Number employed less than 35 hours.

# Exploring Relationships in the Data Through Scatter Plots

Next, we'll use scatter plots to explore some relationships between variables in the dataset.

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Number of Each Major and Unemployment Rate')
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Women as Share of Total and Unemployment Rate')
```

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter', title='Number of Male Graduates in the Major and Median Salary')
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter', title='Number of Women Graduates in the Major and Median Salary')
```

## Observations

### Popular Majors and Salary

The first chart indicates that students in more popular majors don't make more money. As the number of students in a particular major goes up the trend of salary earnings generally goes down. While there are some examples of less popular majors that have low earnings compared to popular majors, on the whole it appears that many less popular majors have substantially higher earnings that popular majors. Perhaps it is a matter of assuring that one's choice of a less popular major is also associated with a higher-in-demand job skill. (We all know of some majors, after all, that are rarely chosen, but also not in demand by potential employets.)

### Salary and Male/Female Representation in Major

As we study the fifth and sixth charts containing data on the share of female/male students in the major compared to median salary, we can see that as more an more men are in the overall percentage of the majors, the median earnings gradually taper down and flatten out around $40,000.

When more and more women are represented in the major, the meadian earning taper more quickly and flatten out around $35,000.

### Salary and Number of Full-Time Employees

As the number of full-time employees represented in a major increases, then the salary generally becomes more predictable at near $40,000. Interestingly, fewer full-time employees means a much greater range of salaries, so in some cases this may relate to the rarity of a job skill leading to a higher salary for a major. In other cases, more part-time workers in a major seems to associate with lower overall earnings than if a major was represented by graduates mostly working full-time.
